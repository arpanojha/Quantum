```python
from qiskit import *
from qiskit.circuit.library import MCXGate
```

```python
import qiskit.quantum_info as qi
from qiskit.circuit.library import FourierChecking
from qiskit.visualization import plot_histogram
```

```python
qr_0 = QuantumRegister(1,'c')
qr_u = QuantumRegister(1,'u')
qr_v = QuantumRegister(1,'v')
#qr = QuantumRegister(3)
cr = ClassicalRegister(3)
circuit = QuantumCircuit(qr_0,qr_u,qr_v,cr)
%matplotlib inline
```

```python
circuit.draw(output='mpl')
```

```python
#circuit.x(2)
#circuit.x(0)
circuit.h(0)
gate = MCXGate(2)
circuit.append(gate, [1,2,0])
circuit.h(0)
circuit.draw('mpl')
```

```python
circuit.measure(qr_0,cr[0])
circuit.measure(qr_u,cr[1])
circuit.measure(qr_v,cr[2])
circuit.draw('mpl')
```

```python
simulator = Aer.get_backend('qasm_simulator')
result = execute(circuit,backend=simulator).result()
from qiskit.tools.visualization import plot_histogram
plot_histogram(result.get_counts(circuit))
```
