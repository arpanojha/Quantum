```python
from qiskit import *
import matplotlib.pyplot as plt
import numpy as np
```

```python
qr = QuantumRegister(2)
cr = ClassicalRegister(2)
circuit = QuantumCircuit(qr,cr)
%matplotlib inline
circuit.draw(output='mpl')
```

```python
circuit.cx(1,0)
circuit.draw(output='mpl')
```
