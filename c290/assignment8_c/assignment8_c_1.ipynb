```python
from qiskit import *
import matplotlib.pyplot as plt
import numpy as np
```

```python
qr = QuantumRegister(3)
cr = ClassicalRegister(3)
circuit = QuantumCircuit(qr,cr)
%matplotlib inline
circuit.draw(output='mpl')
```

## Desired CNOT

```python
circuit.cx(0,2)
circuit.draw(output='mpl')
```

## Which one is correct?

```python
qr_1 = QuantumRegister(3)
cr_1 = ClassicalRegister(3)
circuit = QuantumCircuit(qr_1,cr_1)
%matplotlib inline
circuit.draw(output='mpl')
```

```python
circuit.cx(1,2)
circuit.cx(0,1)
circuit.cx(1,2)
circuit.cx(0,1)
circuit.cx(1,2)
circuit.draw(output='mpl')
```
